```python
# 1. Attached as a photo-done manually

#2.
#function to calculate number of iterations:
def calculateMaxima(b,e,c,m,R,x):
    while abs(grad(x))>e and c!=b:
    
        if f(x)==m:
            break
        x=x+R*grad(x)
        c=c+1
        print('x=',x)
        print('f(x)=',f(x))
        print('g(x)=',grad(x))
        
    print('The function took ',c,' number of iterations')


## Define the function and gradient
def f(x):
    return -x**2
def grad(x):
    return (-2*x)
## see how it works
x0 = 2
R=0.3
c=0
e=1e-6
m = 0
b=1000
print('f(x)=',f(x0))
print('g(x)=',grad(x0))
calculateMaxima(b,e,c,m,R,x0)


```

```python
#Normal Distribution
#number of repetitions R:
R=1000
def Normal(N):
    a=np.random.normal(size=(N,))
    mu = np.random.normal(size=(N,R)).mean(axis=1)
    #mean of means
    print('mean of means is: ',mu.mean())
    #standard deviation
    print('Standard deviation is: ',a.std())
    #confidence interval
    print("Confidence Interval is : ",a.mean()-(1.96*a.std())," to ",a.mean()+(1.96*a.std()))

#Repeat the above with three different N:
Normal(10)
Normal(100)
Normal(10000)



def meansamples(m, n):
    mu = np.random.normal(size=(m,n)).mean(axis=1)
    barx = mu.mean()
    std = np.sqrt(mu.var())
    x1 = barx - 1.96*std
    x2 = barx + 1.96*std
    inside = np.logical_and(mu >= x1, mu <= x2).mean()
    return (x1, x2, inside)

meansamples(10,1000)  

meansamples(100,1000)

meansamples(10000,1000)


```


Means of means moves closer to zero with increase in the value of N(considering the absolute value because distribution is on both sides positive and negative). As we increase the sample size, the distribution gets more and more concentrated towards the mean. In other words, the standard deviation keeps on decreasing.
the confidence interval seems to be increasing in width based on the size of N.
Also, we can see that confidence interval is fairly equally distributed on the both sides of 0.

```python
#Pareto:

def Pareto(N):
    R=1000
    shape, mode = 10.,2.
    values =(np.random.pareto(shape,size=(N,)) +1)* mode
    
    means = ((np.random.normal(shape, size=(N,R))+1)*mode).mean(axis=1)
    
    #calculate the mean of means
    print('Mean of means =',means.mean())
    #calculate sd
    print('Std dev =  ',values.std())
    #CI
    x1 = values.mean()-(1.96*values.std())
    x2 = values.mean()+(1.96*values.std())
    print("Conf interval = " + str(x1) + " : " +str(x2))
    
Pareto(10)
Pareto(100)
Pareto(1000)
```

The value of mean does not vary much with change in the value of N.
There is no definite pattern observed in the standard deviation, however ideally it should increase.

More extreme events may occur often in pareto distribution compared to random distribution.
